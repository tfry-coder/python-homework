```python
# import pandas first!
import pandas as pd
```

```python
#load csvs
listingsdf = pd.read_csv('menu_data.csv')
listingsdf.shape
```

```python
listingsdf.head()
```

```python
#load sales data
sales = pd.read_csv('sales_data.csv')

sales.head()
```

```python
#how many rows
len(sales)
```

```python
#rename columns for merging
listingsdf = listingsdf.rename({'Quantity':'Menu_Item'}, axis='columns')
listingsdf
```

```python
#Combining dataframes
dfjoin = pd.merge(sales,listingsdf, on='Menu_Item', how='inner')
dfjoin
```

```python
#removing unneeded columns
dfjoin = dfjoin.drop(columns=['',''])
dfjoin.head()
```

```python
# create pivot table by product groups
import numpy as np

dfpivot = pd.pivot_table(dfjoin, values=['Quantity', 'Price','COGS'],
                        index=['Product'],aggfunc=np.sum)
dfpivot
```

```python
#calculating profit and making another column for it
dfpivot['Profit'] = (dfpivot['Price'] - dfpivot['COGS'])
dfpivot
```

```python
#export to csv
dfpivot.to_csv('ProductReport.csv')
```
